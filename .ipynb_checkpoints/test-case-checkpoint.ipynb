```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style="darkgrid")
```

## Exploring Data


```python
# Reading dataset
df =  pd.read_csv('data/forestfires.csv')
df.head()
```

```python
# Descriptive statistics
num_cols = df.select_dtypes(include='number').columns
df[num_cols[2:]].describe()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 12))
sns.heatmap(df[num_cols[2:]].corr(), annot=True, cmap="YlGnBu", linewidths=.8)
```

```python
#  Distributions of meteorological feature
fig, axs =  plt.subplots(nrows=2, ncols=2, figsize=(12,12))
axs = np.ravel(axs)

for i, col in enumerate(num_cols[6:10]):
    plt.sca(axs[i])
    sns.histplot(data=df, x=col, kde=True, line_kws={'linewidth':2, 'linestyle':'--'})
plt.tight_layout()
plt.show()
```

```python
#  distribution of burned area
fig, axs =  plt.subplots(nrows=1, ncols=2, figsize=(12,6))

b_area = df['area']

plt.sca(axs[0])
sns.histplot(b_area, kde=True, line_kws={'linewidth':2, 'linestyle':'--'})

# reduced skeweness area 
plt.sca(axs[1])
sns.histplot(np.log(b_area+1), kde=True, line_kws={'linewidth':2, 'linestyle':'--'})
axs[1].set_xlabel("u_area")

plt.tight_layout()
plt.show()
```

```python
# Scatter plot between area and meteorologicalfeatures
fig, axs =  plt.subplots(nrows=2, ncols=2, figsize=(15,15))
axs = np.ravel(axs)

for i, col in enumerate(num_cols[6:10]):
    plt.sca(axs[i])
    scat = sns.scatterplot(x=df[col], y=np.log(df['area']+1))
```

```python
# Data mapping of categorical features

month_map = {'jan':1, 'feb':2, 'mar':3, 
             'apr':4, 'may':5, 'jun':6, 
             'jul':7, 'aug':8, 'sep':9, 
             'oct':10, 'nov':11, 'dec':12}

day_map = {'mon':1, 'tue':2, 'wed':3,
          'thu':4, 'fri':5, 'sat':6, 'sun':7}

df.month = df.month.map(month_map)
df.day = df.day.map(day_map)
```

## Machine Learning Models


```python
# Splitting dataset into training and testing data

target = np.log(df['area']+1)
features = df.drop(columns = 'area')

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 196)

print ("Train data set size : ", X_train.shape)
print ("Test data set size : ", X_test.shape)
```

```python
#  Random Forest Regressor
rfr_model = RandomForestRegressor(n_estimators=300, random_state = 42, n_jobs = -1)
rfr_model.fit(X_train, y_train)

# Predictions
rfr_pred = rfr_model.predict(X_test)

# Scores
print ("Mean Squared Error : ", mean_squared_error(y_test, rfr_pred))
print ("r2 Score : ", r2_score(y_test, rfr_pred))

#Checking the accuracy
rfr_model_accuracy = round(rfr_model.score(X_train, y_train)*100,2)
print(round(rfr_model_accuracy, 2), '%')
```

```python
# Support Vector Regression (SVR)
svr_model = SVR(kernel = 'linear')
svr_model.fit(X_train, y_train)

# Predictions
svr_pred = svr_model.predict(X_test)

# Scores
print ("Mean Squared Error : ", mean_squared_error(y_test, svr_pred))
print ("r2 Score : ", r2_score(y_test, svr_pred))

#Checking the accuracy
svr_model_accuracy = round(svr_model.score(X_train, y_train)*100,2)
print(round(svr_model_accuracy, 2), '%')
```
